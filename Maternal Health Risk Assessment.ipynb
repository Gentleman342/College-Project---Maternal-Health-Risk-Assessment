# Data Fields
* Age :- This column likely represents the age of the pregnant women in the dataset. Age can be an important factor in maternal health risk assessment, as certain age groups may be associated with higher risks for complications during pregnancy.


* SystolicBP :- Systolic blood pressure (SBP) is the highest pressure in the arteries when the heart beats and pushes blood out into the body. Elevated systolic blood pressure can indicate conditions such as hypertension, preeclampsia, or gestational hypertension, which are important risk factors in maternal health.


* DiastolicBP :- Diastolic blood pressure (DBP) is the lowest pressure in the arteries when the heart is at rest between beats. Monitoring diastolic blood pressure, along with systolic blood pressure, helps in assessing overall blood pressure health and identifying potential risks during pregnancy.


* BS (Blood Sugar) :- This column likely represents the blood sugar levels of the pregnant women. Abnormal blood sugar levels, such as in gestational diabetes, can increase the risk of complications for both the mother and the baby during pregnancy and childbirth.


* BodyTemp (Body Temperature) :- Body temperature may be included as a feature in your dataset. Changes in body temperature can sometimes indicate infections or other health issues that may affect pregnancy outcomes.


* HeartRate :- Heart rate refers to the number of times the heart beats per minute. Monitoring heart rate can provide insights into cardiovascular health and potential risks during pregnancy.


* RiskLevel :- This column likely indicates the risk level associated with each set of features for the pregnant women.

# Data Distribution using Pie Chart

```python
import matplotlib.pyplot as plt
from ipywidgets import interact

# Function to create pie charts for selected columns
def plot_pie(column_name):
    plt.figure(figsize=(8, 8), facecolor='white')
    
    category_counts = data[column_name].value_counts()
    top_4_categories = category_counts.head(4)
    other_categories_count = category_counts.iloc[4:].sum() if len(category_counts) > 4 else 0
    
    # Create labels for the pie chart
    labels = list(top_4_categories.index)
    if other_categories_count > 0:
        labels.append('Others')
    
    # Combine counts for top 4 categories and others
    sizes = list(top_4_categories.values)
    if other_categories_count > 0:
        sizes.append(other_categories_count)
    
    plt.pie(sizes, labels=labels, autopct='%.2f%%', startangle=90, textprops={'fontsize': 12})
    plt.title(f'Distribution of {column_name}', fontsize=15)
    plt.show()

# Get the list of columns for which you want to create pie charts
columns_for_pie = ['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate', 'RiskLevel']

# Create an interactive widget with a dropdown menu
interact(plot_pie, column_name=columns_for_pie)
```

* We will remove the outlier in HeartRate column as it is not possible to have 7 heart rate and low risk . And other columns are needed for the better prediction.

## Handling Outliers

```python
from scipy import stats
IQR = stats.iqr(data.HeartRate, interpolation = 'midpoint') 
IQR

## min and max limit calculation

# First quartile (Q1) 
Q1 = data.HeartRate.quantile(0.25)

# Third quartile (Q3) 
Q3 =data.HeartRate.quantile(0.75)

##getting the limits
min_limit=Q1 - 1.5*IQR

max_limit=Q3 +1.5*IQR

print('max_limit for px_height : ',max_limit)
print('min_limit for px_height : ',min_limit)
```

```python
data.loc[data['HeartRate'] < min_limit]
```

```python
data.loc[data['HeartRate'] < min_limit,'HeartRate']=np.median(data.HeartRate)
```

```python
data.loc[data['HeartRate'] < min_limit]
```

# Model Creation

```python
## Creating independent and dependent variable
X=final_data.iloc[:,:-1]
y=final_data.RiskLevel
X
```

```python
y
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
y_train.value_counts()
```

## Smoting Technique

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_train_smote, y_train_smote = smote.fit_resample(X_train,y_train)
```

```python
y_train_smote.value_counts()
```

# Model Evaluation

## Logistic Regression Algorithm

```python
from sklearn.metrics import f1_score,accuracy_score,recall_score,precision_score,classification_report
```

```python

from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train_smote,y_train_smote)
LR_y_train_predict=LR.predict(X_test)
accuracy = accuracy_score(y_test,LR_y_train_predict)
print("Accuracy:",accuracy*100,"%")

classification = classification_report(y_test,LR_y_train_predict)
print("\n",classification)
```

### Cross Validation on Logistic Regression 

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation and compute accuracy scores
scores = cross_val_score(LR, X, y, cv=10, scoring='accuracy')

print("Cross Validation Score of Logistic Regression:", scores.mean()*100,"%")
print("Std:", scores.std())
```

## OneVsRest Classifier Algorithm

```python

from sklearn.multiclass import OneVsRestClassifier
ovr_classifier = OneVsRestClassifier(LR)
ovr_classifier.fit(X_train_smote, y_train_smote)
y_pred = ovr_classifier.predict(X_test)
accuracy = accuracy_score(y_pred,y_test)
print("Accuracy:",accuracy*100,"%")

classification = classification_report(y_test, y_pred)
print("\n",classification)
```

### Cross Validation on OneVsRest Classifier

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation and compute accuracy scores
scores = cross_val_score(ovr_classifier, X, y, cv=10, scoring='accuracy')

print("Cross Validation Score of Logistic Regression:", scores.mean()*100,"%")
print("Std:", scores.std())
```

## SVM (Support Vector Machine) Algorithm

```python
# Support Vector Machine

from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train_smote, y_train_smote)
y_hat = svc.predict(X_test)
accuracy = accuracy_score(y_test, y_hat)
print("Accuracy:",accuracy*100,"%")

classification = classification_report(y_test, y_hat)
print("\n",classification)
```

### Cross Validation on SVC

```python
# Perform cross-validation and compute accuracy scores
scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy')

print("Cross Validation Score of SVC:", scores.mean()*100,"%")
print("Std:", scores.std())
```

## KNN (K-Nearest Neighbor) Algorithm

```python
# Knn 

import warnings
warnings.filterwarnings('ignore')
from sklearn.neighbors import KNeighborsClassifier
error_rate =[]
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_smote, y_train_smote)
    y_new = knn.predict(X_test)
    error_rate.append(np.mean(y_new != y_test))
    
plt.figure(figsize=(8,4))
plt.plot(range(1,11),error_rate, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)

knn1 = KNeighborsClassifier(n_neighbors=2)
knn1.fit(X_train_smote, y_train_smote)
y_new = knn1.predict(X_test)
accuracy = accuracy_score(y_test, y_new)
print("Accuracy:",accuracy*100,"%")

classification = classification_report(y_test, y_new)
print("\n",classification)
```

### Cross Validation on KNN

```python
# Perform cross-validation and compute accuracy scores
scores = cross_val_score(knn1, X, y, cv=10, scoring='accuracy')

# Print the results
print("Cross Validation Score of KNN:", scores.mean()*100,"%")
print("Std:", scores.std())
```
